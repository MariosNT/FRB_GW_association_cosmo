Recalculate all $DM_{MW,ISM}$ based on our (gl, gb) and pygedm (see `ne2001_compare.ipynb`)

```python
import sys
sys.path.append('../../Python_scripts')

## Import packages
from config import *
from support import *
from cosmo_support import *
import seaborn as sns

import pygedm
```

load data

```python
data=pd.read_csv('../../Data/FRB.csv')
data.head()
```

```python
cosmo = Planck18
```

```python
for idx in range(len(data)):
    dis=cosmo.luminosity_distance(data['z'].iloc[idx])/u.Mpc
    data.loc[idx,'DM_MW, ISM(NE2001)'] = pygedm.dist_to_dm(gl=data['gl'].iloc[idx], 
                            gb=data['gb'].iloc[idx], 
                            dist=dis.value*1e6, 
                            method='ne2001')[0].value
    data.loc[idx,'DM_MW, ISM(YMW16)'] = pygedm.dist_to_dm(gl=data['gl'].iloc[idx], 
                            gb=data['gb'].iloc[idx], 
                            dist=dis.value*1e6, 
                            method='ymw16')[0].value
data.head()
```

```python
DM_MW_halo=30
data['DM_ext(ne2001)'] = data['DM']-DM_MW_halo-data['DM_MW, ISM(NE2001)'] #apply(lambda x: f'{x:.2f}')
data['DM_ext(ymw16)'] = data['DM']-DM_MW_halo-data['DM_MW, ISM(YMW16)'] #.apply(lambda x: f'{x:.2f}')
data.head()
```

```python
data['DM_MW, ISM(NE2001)'] = data['DM_MW, ISM(NE2001)'].round(2) #apply(lambda x: f'{x:.2f}')
data['DM_MW, ISM(YMW16)'] = data['DM_MW, ISM(YMW16)'].round(2) #.apply(lambda x: f'{x:.2f}')
data.head()
```

```python
for idx in range(len(data)):
    if np.abs(data['DM_ext(ne2001)'].iloc[idx] - data['DM_ext'].iloc[idx]) > 0.1:
        print(f"{idx}: {data['FRB'].iloc[idx]}: ne2001={data['DM_ext(ne2001)'].iloc[idx]} != paper {data['DM_ext'].iloc[idx]}")
```
