```python
import numpyro
import numpyro.distributions as dist
from numpyro import sample

def model(data):
    mean = sample('mean', dist.Normal(0, 1))
    std = sample('std', dist.Exponential(1))
    with numpyro.plate('data', len(data)):
        sample('obs', dist.Normal(mean, std), obs=data)
```

```python
import numpy as np

np.random.seed(0)
data = np.random.normal(3, 2, size=100)
```

```python
from numpyro.infer import MCMC, NUTS
import jax

nuts_kernel = NUTS(model)
mcmc = MCMC(nuts_kernel, num_warmup=500, num_samples=1000)
mcmc.run(jax.random.PRNGKey(0), data=data)
mcmc.print_summary()
```
