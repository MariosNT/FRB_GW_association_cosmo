```python
# 定义符号变量
x, a, m = sp.symbols('x a m')

# 例1：计算∫(x^m)dx
expr1 = x**m
result1 = sp.integrate(expr1, x)
print("∫(x^m)dx =", result1)

# 例2：含参数的指数函数积分
expr2 = sp.exp(a*x)
result2 = sp.integrate(expr2, x)
print("∫(e^(a*x))dx =", result2)

# 例3：含参数的三角函数积分
expr3 = sp.sin(a*x)
result3 = sp.integrate(expr3, x)
print("∫sin(a*x)dx =", result3)

# 例4：有定义域限制的定积分
expr4 = x**2 + a*x
result4 = sp.integrate(expr4, (x, 0, a))
print(f"∫_0^a (x^2 + a*x)dx =", result4)
```

```python
sigmas=np.linspace(.1, 5, 500)
Vars=np.zeros_like(sigmas)

for i, sigma in enumerate(sigmas):
    C0=C0_sigma(sigma)
    A = find_A_sigma(C_0=C0, sigma=sigma, alpha=3, beta=3)
    Vars[i]=calculate_var(C0=C0, A=A, sigma_DM=sigma)
    
plt.plot(Vars, sigmas**2)
plt.ylabel('sigma_DM^2')
plt.xlabel('Variance')
plt.xscale('log')
plt.yscale('log')
```

```python
def test(var,a,b):
    return a*np.sqrt(var)

x=Vars
x=np.array(x)
y=sigmas
y=np.array(y)

popt, pcov = curve_fit(test, x, y, p0=[1.0, 1.0])

a_fit, b_fit = popt
print(a_fit, b_fit)

y_fit = test(Vars, a_fit, b_fit)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, 
            color='blue', 
            alpha=0.3, label='Data')
plt.plot(x, y_fit, 
         color='green', 
         label=f'$\sigma={a_fit:.2f}*Var**{b_fit:.2f}$')

plt.ylabel('sigma_DM')
plt.xlabel('Variance')
```

```python
def test_int(z):
    Om=OMEGA_MATTER
    def dDc(x):
        return 1/np.sqrt(Om*(1+x)**3+(1-Om))
    
    def dDM(x):
        return (1+x)/np.sqrt(Om*(1+x)**3+(1-Om))
    
    int1,_=quad(dDc, 0, z)
    int2,_=quad(dDM, 0, z)
    
    return int1/int2**2

def find(z,F):
    var=F/z
    sigma=sigma_var(var)
    return sigma

def find_num(z,F):
    num_int=np.vectorize(test_int)
    #var=F/np.sqrt(z)
    var=F*num_int(z)
    sigma=sigma_var(var)
    return sigma

x=df['z']
x=np.array(x)
y=df['sigma']
y=np.array(y)
guess=6.0
popt, pcov = curve_fit(find, x, y, p0=[guess])
popt_num, pcov_num = curve_fit(find_num, x, y, p0=[guess])

a_fit = popt[0]
print(a_fit)
a_fit_num = popt_num[0]
print(a_fit_num)

x_fit=np.linspace(0.1,np.max(x),100)
y_fit = find(x_fit, a_fit)
y_fit_num = find_num(x_fit, a_fit_num)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, 
            color='blue', 
            alpha=0.3, label='$\sigma-z$ relation based on IllustrisTNG (Z. J. Zhang et al. 2021)')
plt.plot(x_fit, y_fit, 
         color='green', 
         label=f'fit: \sigma=\sigma(Var), Var={a_fit:.2f}/z')
plt.plot(x_fit, y_fit_num, 
         color='red', 
         label=f'fit: \sigma=\sigma(Var), Var={a_fit:.2f}*f_num')

plt.xlabel('z')
plt.ylabel('$\sigma$')
plt.legend()
plt.grid(True)
plt.show()
```
